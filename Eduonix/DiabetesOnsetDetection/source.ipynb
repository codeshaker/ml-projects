```python
import pandas as pd
import numpy as np

# Load the dataset downloaded from kaggle

names = ['n_pregnant', 'glucose_concentration', 'blood pressure (mm Hg)', 'shin thickness (mm)', 'serum_insulin (mu U/ml)',
        'BMI', 'pedigree_function', 'age', 'class']

df = pd.read_csv('pima-indians-diabetes.csv', names = names)
```

```python
# Preprocess the dataset, mark zero values as NaN and drop

columns = ['glucose_concentration', 'blood pressure (mm Hg)', 'shin thickness (mm)', 'serum_insulin (mu U/ml)',
        'BMI']

#Replacing the 0 values in these columns with NaN
for col in columns:
    df[col].replace(0, np.NaN, inplace=True)
    
df.describe()
```

```python
# Drop the rows with missing values
df.dropna(axis = 0, inplace = True)

df.describe()
```

```python
# Conver the df to numpy array
dataset = df.values

print(dataset.shape)
```

```python
# Split that into input (X) and output (Y)

X = dataset[:,0:8]
Y = dataset[:,8].astype(int)
```

```python
print(X.shape)
print(Y.shape)
print(Y[:5])
```

```python
# Normalize the data using sklearn standdardScalar
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X)
```

```python
print(scaler)
```

```python
# Transform and display the training data
X_standardized = scaler.transform(X)

data = pd.DataFrame(X_standardized)
data.describe()
```

```python
# Import necessary packages
from sklearn.model_selection import GridSearchCV, KFold
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.optimizers import Adam
```

```python
# Start defining the models

def create_model():
    #create model
    model = Sequential()
    model.add(Dense(8, input_dim = 8, kernel_initializer='normal', activation='relu'))
    model.add(Dense(4, input_dim = 8, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    # compile the model
    adam = Adam(lr = 0.01)
    model.compile(loss= 'binary_crossentropy', optimizer=adam, metrics=['accuracy'])
    return model

model = create_model()
print(model.summary())
# 
```

```python
#Define random seed -for same initialization of data.
seed=6
np.random.seed(seed)

# Start defining the models

def create_model():
    #create model
    model = Sequential()
    model.add(Dense(8, input_dim = 8, kernel_initializer='normal', activation='relu'))
    model.add(Dense(4, input_dim = 8, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    # compile the model
    adam = Adam(lr = 0.01)
    model.compile(loss= 'binary_crossentropy', optimizer=adam, metrics=['accuracy'])
    return model

# Create a model
model = KerasClassifier(build_fn = create_model, verbose = 1)

# Define the grid search parameters
batch_size = [10, 20, 40]
epochs = [10, 50, 100]

# make a dictionary for above search parameters
param_grid = dict(batch_size=batch_size, epochs=epochs)

# Build and fit the GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv = KFold(random_state=seed), verbose = 10)

grid_result = grid.fit(X_standardized, Y)

#Summarize the results
print("Best: {0}, using {1}".format(grid_result.best_score_,grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']

for mean,stdev,param in zip(means,stds,params):
    print('{0} ({1}) with: {2}'.format(mean, stdev, param))
```

```python
# Import necedssary packages
from keras.layers import Dropout

#Define random seed -for same initialization of data.
seed=6
np.random.seed(seed)

# Start defining the models

def create_model(learn_rate, dropout_rate):
    #create model
    model = Sequential()
    model.add(Dense(8, input_dim = 8, kernel_initializer='normal', activation='relu'))
    # Prevent a neuron from becoming too heavy weighted or important.
    model.add(Dropout(dropout_rate))
    model.add(Dense(4, input_dim = 8, kernel_initializer='normal', activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1, activation='sigmoid'))

    # compile the model
    adam = Adam(lr = learn_rate)
    model.compile(loss= 'binary_crossentropy', optimizer=adam, metrics=['accuracy'])
    return model

# Create a model (Already found best grid seach params above so using it here itself)
model = KerasClassifier(build_fn = create_model, epochs = 100, batch_size = 40, verbose = 0)

# Define the grid search parameters
learn_rate = [0.001,0.01,0.1]
dropout_rate = [0.0, 0.1, 0.2]

# make a dictionary for above search parameters
param_grid = dict(learn_rate=learn_rate, dropout_rate=dropout_rate)

# Build and fit the GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv = KFold(random_state=seed), verbose = 1)

grid_result = grid.fit(X_standardized, Y)

#Summarize the results
print("Best: {0}, using {1}".format(grid_result.best_score_,grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']

for mean,stdev,param in zip(means,stds,params):
    print('{0} ({1}) with: {2}'.format(mean, stdev, param))
```

```python
# Import necedssary packages


#Define random seed -for same initialization of data.
seed=6
np.random.seed(seed)

# Start defining the models

def create_model(neuron1, neuron2):
    #create model
    model = Sequential()
    model.add(Dense(neuron1, input_dim = 8, kernel_initializer='normal', activation='linear'))
    # Prevent a neuron from becoming too heavy weighted or important.
    model.add(Dropout(0.1))
    model.add(Dense(neuron2, input_dim = neuron1, kernel_initializer='normal', activation='linear'))
    model.add(Dropout(0.1))
    model.add(Dense(1, activation='sigmoid'))

    # compile the model
    adam = Adam(lr = 0.01)
    model.compile(loss= 'binary_crossentropy', optimizer=adam, metrics=['accuracy'])
    return model

# Create a model (Already found best grid seach params above so using it here itself)
model = KerasClassifier(build_fn = create_model, epochs = 100, batch_size = 40, verbose = 0)

# Define the grid search parameters
neuron1 = [4, 8, 16]
neuron2 = [2, 4, 8]

# make a dictionary for above search parameters
param_grid = dict(neuron1=neuron1, neuron2=neuron2)

# Build and fit the GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv = KFold(random_state=seed),refit= True, verbose = 1)

grid_result = grid.fit(X_standardized, Y)

#Summarize the results
print("Best: {0}, using {1}".format(grid_result.best_score_,grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']

for mean,stdev,param in zip(means,stds,params):
    print('{0} ({1}) with: {2}'.format(mean, stdev, param))
```
